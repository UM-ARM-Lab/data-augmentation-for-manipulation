How does loss change with latent size in the linear case?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.array([1,  2, 3, 4, 5, 6, 16])
losses = np.array([302.3, 49.35, 38.03, 34.83, 32.62, 30.52, 34.38])
plt.plot(m,losses)
plt.xlabel("Size of Latent Space (M)")
plt.ylabel("Loss after 30k epochs")
plt.show()
```

```python
plt.plot(m[1:],losses[1:])
plt.xlabel("Size of Latent Space (M)")
plt.ylabel("Loss after 30k epochs")
plt.show()
```
