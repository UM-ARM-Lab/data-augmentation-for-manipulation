# Checking for Distribution Mismatch

```python
from __future__ import print_function
from link_bot_notebooks import notebook_finder
from link_bot_notebooks import toy_problem_optimization_common as tpo
import numpy as np
from math import acos, atan2
from scipy import stats
from itertools import cycle
import matplotlib.pyplot as plt
np.set_printoptions(precision=6)
```

```python
goal = np.zeros((6,1))
```

```python
def plot_distribution(data):
    angles = np.ndarray(data.shape[2]*data.shape[0])
    for i in range(data.shape[2]):
        d = data[0, :, i]
        u = atan2(d[3] - d[1], d[2] - d[0])
        v = atan2(d[7], d[6])
        angle = u - v
        angles[i] = angle
    counts, bins, _ = plt.hist(angles, bins=20)
    plt.xlabel("angle between link 0 and direction of head velocity")
    plt.ylabel("occurances")
    # D, p = stats.kstest(counts, 'uniform')
    # print("KS Test D={} p={}".format(D, p))
```

```python
dataset = "../../../link_bot_teleop/data/straight_pulls.txt"
data = tpo.load_train(dataset, n_steps=0, N=6, L=2, extract_func=tpo.link_pos_vel_extractor2(6))
print(data.shape[2], 'samples')
plot_distribution(data)
```
