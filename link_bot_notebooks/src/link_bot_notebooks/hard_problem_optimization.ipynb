```python
def plot_one_step(model, data, dt, g, first_n=100):
    actions = [d[1] for d in data]
    states = [d[0] for d in data]
    plt.figure(figsize=(10,10))
    o_1 = []
    o_2 = []
    for s, u in list(zip(states, actions))[:first_n]:
        o = model.reduce(s)
        o_1.append(o[0,0])
        o_2.append(o[1,0])
        c_hat = model.cost_of_o(o, g)
        o_ = model.predict_from_o(o, u, dt)
        plt.plot([o[0,0], o_[0,0]], [o[1,0], o_[1,0]], color='r')
        
    plt.plot(o_1, o_2)
    plt.xlabel("time steps")
    plt.ylabel("o")
    plt.axis('equal')
    plt.show()

plot_one_step(ccp_model, data, dt, g)
```
